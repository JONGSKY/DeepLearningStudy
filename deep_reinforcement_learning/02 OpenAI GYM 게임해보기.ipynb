```python
import gym
```

```python
env = gym.make("FrozenLake-v0")
observation = env.reset()
```

```python
for _ in range(1000):
    env.render() # 환경을 화면에 출력
    action = env.action_space.sample() # 환경에 따라 적절한 action을 선정
    observation, reward, done, info = env.step(action)
#     print(done)
```
